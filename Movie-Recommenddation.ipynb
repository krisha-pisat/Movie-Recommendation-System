```python
 # as it a list of strings we cannot print the values of key 'name'
# so there is a module in python called as ast
#And a method in ast called as ast.literal_eval() which converts the string into a list
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# We will have too convert the text in the Tag feature into Vectors 
# This process is called as Text Vectorisation
# The technique we use for this is called as Bag of Words
# There are other techniques as well like 1) tfidf 2)wordevec
```

```python
# Step 1) Concatenate all the Tag Features of every Row [Concan_Tags]
# Step 2) Find 5000 most used words
# Step 3) extract those most used words
# Step 4) Choose the Tag of any 1 movie and calculate the frequency of all the words in the Tag eg)'Action' or 'frequency'
#  |W1|W2|W3|W4|........|W5000|
#  |2 |3 |0 |1|..........|2435|  ---> this is a vector now
# Now we have a dataframe of the frequencies of each word in a row
# a dimension of 5000 is created , so it is considered a good practice if u use a lower number of dimension
```

```python
# When we are performing vectorisation , dont consider 'STOP' words
# STOP words mean which are used for sentence formation but do not have any
# contribution in the meaning of the sentence for example)in,are,is,or,and, from
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```
